```python
import pandas as pd

df = pd.read_csv("data/모의고사 2회/titanic.csv")

cate_type = ["sex", "embarked"]
cate_null_cols = []

number_type = ["pclass", "age", "sibsp", "parch", "fare"]
number_null_cols = []

for col in number_type:
    null_col_cnt = len(df[(df[col].isnull()==True)])
    if null_col_cnt > 0:
        number_null_cols.append(col)

for col in number_null_cols:
    this_median = df[col].median()
    df[col].fillna(this_median, inplace=True)

for col in cate_type:
    null_col_cnt = len(df[(df[col].isnull()==True)])
    if null_col_cnt > 0:
        cate_null_cols.append(col)
#         print(col, null_col_cnt)

for col in cate_null_cols:
    this_mode = df[col].mode()
    df[col].fillna(this_mode[0], inplace=True)

age_1_crit = [
    [0, 0, 10],
    [1, 10, 20],
    [2, 20, 30],
    [3, 30, 40],
    [4, 40, 50],
    [5, 50, 60],
    [6, 60, 70],
    [7, 70, 80],
    [8, 80, 90]
]
df["age_1"] = ""
for crit in age_1_crit:
    df.loc[(df["age"] >= crit[1])&(df["age"] < crit[2]), "age_1"] = crit[0]
    
df
```

#### 데이터 전처리 및 훈련/테스트 셋 분할

```python
from sklearn.model_selection import train_test_split

df["sex_1"] = ""
df.loc[(df["sex"] == "male"), "sex_1"] = 1
df.loc[(df["sex"] == "female"), "sex_1"] = 2

df["embarked_1"] = ""
df.loc[(df["embarked"] == "S"), "embarked_1"] = 1
df.loc[(df["embarked"] == "Q"), "embarked_1"] = 2
df.loc[(df["embarked"] == "C"), "embarked_1"] = 3

ind_vars = ["pclass", "sex_1", "sibsp", "parch", "fare", "embarked_1"]
dep_var = ["survived"]
X_train, X_test, y_train, y_test = train_test_split(df[ind_vars], df[dep_var], test_size=0.3, train_size=0.7, random_state=12345)
```

#### 의사결정나무

```python
#의사결정나무
import operator
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

# dtc = DecisionTreeClassifier(criterion="entropy") #에러율 25%
# dtc = DecisionTreeClassifier(criterion="gini") #에러율 23%
result_list = {}
for i in range(2, 21):
    dtc = DecisionTreeClassifier(criterion="gini", max_depth=5, max_leaf_nodes=i) #에러율 23%
    fit = dtc.fit(X_train, y_train)
    predict = fit.predict(X_test)
    prd_success_rate = fit.score(X_test, y_test)
    result_list[i] = prd_success_rate
#     print("노드갯수 : {}개, 예측정확도 : {:.4f}".format(i, prd_success_rate))

sdic = sorted(result_list.items(), key=lambda x: x[1], reverse=True)
dtc = DecisionTreeClassifier(criterion="gini", max_depth=5, max_leaf_nodes=sdic[0][0])
fit = dtc.fit(X_train, y_train)
print("지니 지표 사용, 노드갯수 {}개일 때 예측정확도 {:.2f}%로 가장 높음".format(sdic[0][0], sdic[0][1]*100))
plt.figure(figsize=(15,10))
plot_tree(fit, filled=True, proportion=True)
```

#### 랜덤 포레스트

```python
#참고 : https://m.blog.naver.com/samsjang/220979751089
import operator
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

result_list = {}
for i in range(2, 21):
    rfc = RandomForestClassifier(n_estimators=200, criterion="entropy", max_depth=5, max_leaf_nodes=i, oob_score=True)
    fit = rfc.fit(X_train, y_train["survived"])

    predict = fit.predict(X_test)
    prd_success_rate = fit.score(X_test, y_test)
    result_list[i] = prd_success_rate
    print("노드갯수 : {}개, OOB : {:.4f}, 예측정확도 : {:.4f}".format(i, fit.oob_score_, prd_success_rate))

sdic = sorted(result_list.items(), key=lambda x: x[1], reverse=True)
print("엔트로피 지표 사용, 노드갯수 {}개일 때 예측정확도 {:.2f}%로 가장 높음".format(sdic[0][0], sdic[0][1]*100))
```
